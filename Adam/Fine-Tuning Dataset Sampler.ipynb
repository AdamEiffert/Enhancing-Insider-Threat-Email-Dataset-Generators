```python
import pandas as pd
```

```python
df = pd.read_csv('../data/complete_enron.csv')
```

The following code converts all the Enron emails into the required format for fine-tuning. It then samples 2,000 for the training dataset and 500 for the testing dataset. 

```python
emails = []
for i in range(len(df)):
    emails.append('<|endoftext|>\n' + 'Topic: ' + str(df['Topic'][i]) + '\n' + 'Sentiment: ' + df['vader_sentiment_bin'][i] + '\n' + 'Email: ' + str(df['content'][i]) + '\n<|endoftext|>\n')
full = pd.DataFrame(emails,columns=['text'])
training_df = full.sample(n=2000)
testing_df = full.sample(n=500)
st = ''
for i in range(len(training_df)):
    st = st + training_df.reset_index()['text'][i]
training_df.to_csv('../data/train.csv')
testing_df.to_csv('../data/test.csv')
full.to_csv('../data/full_train.csv')
f = open('../data/train.txt', 'w')
f.write(st)
```
